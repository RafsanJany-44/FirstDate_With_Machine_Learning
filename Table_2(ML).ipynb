```python
cel_data = [90, 45, -25, 130, 35, 29, -30, 27, 56, 3, 212, 121, 180, -40, -35]
```

```python
far_data = [194.0, 113.0, -13.0, 266.0, 95.0, 84.2, -22.0, 80.6, 132.8, 37.4, 413.6, 249.8, 356.0, -40.0, -31.0]
```

**Making A DataFrame**

```python
import numpy as np
import pandas as pd


data = {
    "Celsius":cel_data,
    "Fahrenheit":far_data
}
df = pd.DataFrame(data)
```

```python
from sklearn.linear_model import LinearRegression                 #model for finding the pattern
X = df["Celsius"].values.reshape(-1, 1)                           #input
Y = df["Fahrenheit"].values.reshape(-1, 1)                        #output
reg = LinearRegression().fit(X, Y)                                #Train the model to learn the pattern
```

```python
slope = reg.coef_[0]
intercept = reg.intercept_
```

```python
print("Slope:",slope)
print("Intercept:",intercept)
```
